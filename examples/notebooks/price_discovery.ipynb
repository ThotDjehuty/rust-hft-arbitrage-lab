# Price Discovery\n

```python
import requests, pandas as pd, numpy as np
BASE='http://mock_apis:8000'
mids_b=[]; mids_k=[]
for _ in range(60):
    b = requests.get(f'{BASE}/api/binance/orderbook/BTCUSDT').json()
    k = requests.get(f'{BASE}/api/kraken/orderbook/BTCUSD').json()
    mids_b.append((b['bids'][0][0]+b['asks'][0][0])/2)
    mids_k.append((k['bids'][0][0]+k['asks'][0][0])/2)
df = pd.DataFrame({'binance':mids_b,'kraken':mids_k})
print('corr binance(t), kraken(t+1)=', df['binance'].corr(df['kraken'].shift(-1)))
```
